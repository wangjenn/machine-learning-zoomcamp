```python
# 2) How many distinct brands? 

distinct_brands = laptop['Brand'].nunique()
print(distinct_brands)
```

```python
# 3) How many columns have missing values?

missing_values = laptop.isnull().sum()
columns_missing_values = missing_values[missing_values > 0].count()
print(columns_missing_values)
```

```python
# 4) What's the max price of Dell notebooks in dataset? 

dell_max_price = laptop[laptop['Brand'] == 'Dell']['Final Price'].max()
print(dell_max_price)
```

```python
#6) What's the median value of screen?
med_screen = laptop['Screen'].median()
print(med_screen)
```

```python
# 6b). Most frequent value of the screen size?
most_freq_screen = laptop['Screen'].mode()
print(most_freq_screen)

```

```python
#6c). Use fillna method to fill the missing values in Screen column with the most frequent value from the previous step.

fillna_screen = laptop['Screen'].fillna(most_freq_screen)
```

```python
new_med_screen = fillna_screen.median()
print(new_med_screen)
```

```python
# 7) Sum of weights 

df = laptop
# Step 1: Select all "Innjoo" laptops from the dataset
innjoo_laptops = df[df['Brand'] == 'Innjoo']

# Step 2: Select only the columns RAM, Storage, Screen
X = innjoo_laptops[['RAM', 'Storage', 'Screen']].values  # Get underlying NumPy array

# Step 3: Compute matrix-matrix multiplication between the transpose of X and X
XTX = X.T @ X

# Step 4: Compute the inverse of XTX
XTX_inv = np.linalg.inv(XTX)

# Step 5: Create the array y
y = np.array([1100, 1300, 800, 900, 1000, 1100])

# Step 6: Multiply the inverse of XTX with the transpose of X, and then multiply by y
w = XTX_inv @ X.T @ y

# Step 7: Compute the sum of all elements of w
result = w.sum()

print(result)
```
